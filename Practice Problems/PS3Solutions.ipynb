```python
import numpy as np

# Generate 50 unique integers between 10 and 100
arr = np.random.choice(range(10, 101), size=50, replace=False)
print("Input:\n",arr)

i, j = 0, 49
while(i<j):
    arr[i], arr[j] = arr[j], arr[i]
    i+=1
    j-=1

print("Output:\n",arr)
```

```python
import numpy as np

def com_ele(A,B):
    set_a = set(A)
    set_b = set(B)
    return list(set_a & set_b)

A = np.random.randint(1, 10, size=10)
B = np.random.randint(1, 10, size=10)
print(A,B)
print(np.array(com_ele(A,B)))
```
