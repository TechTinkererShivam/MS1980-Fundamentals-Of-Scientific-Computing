```python
a = np.arange(11, dtype=np.int32)
print(f"a = {a}, data-type = {a.dtype}")
b = np.arange(10,21,dtype=np.float64)
print(f"b = {b}, data-type = {b.dtype}")
add = a + b
print(f"add = {add}, data-type = {add.dtype}")
```

NumPy promotes the data type to float64 because:

* **Type Promotion Rule**: When performing operations between arrays of different types (e.g., integers and floats), NumPy promotes the result to the higher precision type to avoid loss of information. In this case, floats have higher precision than integers, so the resulting array is of type float64.

* Row-major order (C-style, order='C') means that the array elements are stored row by row (by default). 
* Column-major order (Fortran-style, order='F') means that the elements are stored column by column.

```python
arr_c = np.arange(6).reshape(2,3,order='C')
print("arr_c =\n",arr_c)
arr_f = np.arange(6).reshape(2,3,order='F')
print("arr_f =\n",arr_f)
```

* To check if a NumPy array is stored contiguously in memory, use the flags attribute: C_CONTIGUOUS for row-major (C-style) and F_CONTIGUOUS for column-major (Fortran-style) storage.

```python
print(arr_c.flags['C_CONTIGUOUS']) # True
print(arr_c.flags['F_CONTIGUOUS']) # False
print(arr_f.flags['F_CONTIGUOUS']) # True
print(arr_f.flags['C_CONTIGUOUS']) # False
```

```python
arr = np.arange(1,7).reshape(2,3)
print(arr)
print(arr.ravel()) # Returns a flattened view (whenever possible), more memory efficient.
print(arr.flatten()) # Returns a flattened copy, no impact on the original array.
print(arr.reshape(-1)) # Similar to ravel(), returns a view when possible but more versatile for reshaping.
```

Note: ravel() always flattens the array to 1D, while reshape(-1) can flatten it or reshape it to any other shape by using -1 for automatic dimension inference.

Structured arrays in Python (NumPy) allow storage of heterogeneous data types in one array. Each element has named fields, like records in databases, with each field holding a different data type (e.g., integers, floats, strings).

```python

# Define a structured data type
data_type = [('name', 'U10'), # U10: Unicode string of max length 10
             ('age', 'i4'), # i4: 32-bit integer
             ('weight', 'f4')] # f4: 32-bit float

# name, age, weight -> fields

# Create a structured array
data = np.array([('Alice', 25, 55.0), ('Bob', 30, 85.5)], dtype=data_type)

# Access the data
print(data) # 
print(data['name']) # ['Alice' 'Bob']
print(data['age'])
print(data['weight'])
```

```python
arr = np.zeros((3,3))
print(arr)
arr += np.ones((3,3)) # in-place addition
print(arr)
arr = arr + np.ones((3,3)) # Not in-place addition
```

The in-place addition (+=) is generally faster than the assignment method (arr + np.ones(...)) because:
* It modifies the existing array directly, avoiding the need to allocate new memory.
* It has less overhead since it does not need to handle a temporary result.

"overhead of creating new arrays" refers to the extra time and resources needed to allocate and initialize new memory for the array, copy data, and potentially manage memory cleanup, all of which can slow down performance and increase memory consumption compared to in-place operations.

```python
arr = np.arange(10)
idx = [2, 5, 8] # fancy indexing
arr[idx] = 0
print(arr)
```

Fancy Indexing in NumPy allows you to access and modify specific elements of an array using arrays or lists of indices.

How Fancy Indexing Works?
* Array Creation: The array arr is created with values ranging from 0 to 9.
* Index Specification: The list idx specifies which indices to access (2, 5, and 8).
* Modification: The elements at these indices in arr are set to 0.
* Result: The array now has the values [0, 1, 0, 3, 4, 0, 6, 7, 0, 9].

Fancy Indexing can lead to unexpected results if the behavior regarding copies vs. views, data types, or non-unique indices is not fully understood.

```python
arr = np.array([1, 2, 3])
arr = np.append(arr, [4, 5])
print(arr)
```

$ Ques: $
Why is this approach inefficient?
* **New Array Creation:** np.append() creates a new array with the original and appended elements.
* **No Shared Memory:** The original array remains unchanged; arr now references the new array.
* **Data Copying:** Elements from both arrays are copied, leading to additional memory allocation.

$ Optimization: $ 
By preallocating memory, using lists for dynamic data collection, utilizing np.concatenate(), or resizing judiciously, we can significantly improve the efficiency of appending elements in NumPy, reducing the overhead of repeated memory allocation and copying.

```python
# Using datetime64[D] enables efficient date manipulations while maintaining a specific precision level (in this case, days).
# The [D] indicates that the precision is at the day level, meaning the array will hold dates without any time component (hours, minutes, seconds).
dates = np.array(['2005-02-22', '2024-10-16'], dtype='datetime64[D]') 
print(dates)
date_diff = dates[1] - dates[0]
print("Difference in days:", date_diff)  # Output: Difference in days:  7176 days

print(dates + np.timedelta64(7, 'D'))
```

* np.timedelta64(7, 'D'): This creates a time delta of 7 days. 
* The 'D' specifies that the unit of the time delta is days.
* A time delta refers to a duration of time.
