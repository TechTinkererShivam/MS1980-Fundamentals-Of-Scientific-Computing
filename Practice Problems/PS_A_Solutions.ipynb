```python
arr = np.arange(5)
print(arr)
view = arr[:3]
view *= 2
print(arr)

# After performing an in-place operation on view, arr has also been modified.
```

```python
# view of a slice (by-default)
arr = np.arange(5)
print("arr = ",arr)
view = arr[:3]
view *= 10
print("arr = ",arr)
```

- On modifying view, the arr has also been modified.

```python
# copy of a slice
arr = np.arange(5)
print("arr = ",arr)
copy = arr[:3].copy()
copy *= 10
print("arr = ",arr)
```

- On modifying copy, the arr has not modified.

```python
arr = np.arange(10,51,10)
print("arr = ",arr)
```

```python
# Integer Indexing
int_indexed  = arr[[1,2,3]] # selects elements at the specified indices
print(int_indexed) # creates a new copy, changes to the new array do not affect the original array.
```

```python
# Slicing
sliced = arr[1:4] # selects a range of elements (from index 1 to index 3)
print(sliced) # modifying the sliced array will also modify the original array because they share the same memory.
```
