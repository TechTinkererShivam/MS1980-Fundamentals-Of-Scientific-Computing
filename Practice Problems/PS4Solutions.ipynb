```python
import numpy as np

r = 1 # radius
n,N = 0, 50000
for i in range(N):
    c = np.random.rand(2) # coordinates(x,y)
    x,y = c[0], c[1]
    if((x**2 + y**2)<=r):
        n += 1

ans = (4*n)/N
print(ans)        

# Yes, the answer is close to PI
```
