* Nth Fibonacci Term Code using Recursion 

```python
def r_fibonacci(n):
    if(n==1):
        return 0
    if(n==2):
        return 1
    return r_fibonacci(n-1)+r_fibonacci(n-2)

# Test Function
print(r_fibonacci(5))
print(r_fibonacci(10))
```

* Nth Fibonacci Term Code without using Recursion 

```python
def l_fibonacci(n):
    if(n==1):
        return 0 # 1st term
    if(n==2):
        return 1 # 2nd term
    t1, t2 = 0, 1
    for i in range(3,n+1):
        t = t1 + t2
        t1 = t2
        t2 = t

    return t
    
# Test Function
print(l_fibonacci(5))
print(l_fibonacci(10))
print(l_fibonacci(15))
```

* code to find 2nd largest number (2-pass solution)

```python
def second_largest(A):
    fl,sl = A[0], A[0] # fl => first largest & sl => second largest
    # loop for first largest
    for i in range(1,len(A)):
        if(A[i]>fl):
            fl = A[i]
    # loop for second largest
    for i in range(1,len(A)):
        if(A[i]>sl and A[i]!=fl):
            sl = A[i]

    if(fl==sl):
        sl = None        

    return sl

# Test Function
print(second_largest([3,1,4,1,5,9,2]))
print(second_largest([1,1,1]))
```
