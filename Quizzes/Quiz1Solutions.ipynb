```python
def list_contains_empty_tuple(mylist_of_tuples):
    for t in mylist_of_tuples:
        if not t:
            return True
    return False

# Test function
my_list1 = [(), (1,2,3),('a',2), ('Hello',),(1.2,0.2,3),()]
print(list_contains_empty_tuple(my_list1)) # Returns True
```
